2.  How does broadcasting work in NumPy ?

- Broadcasting in NumPy is a powerful feature that allows you to perform operations on arrays of different shapes without explicitly reshaping them. It’s like NumPy saying, “I’ll stretch the smaller array to match the bigger one—without actually copying data.”

 What Is Broadcasting?
Broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations. The smaller array is broadcast across the larger array so that they have compatible shapes.

 Broadcasting Rules
NumPy compares arrays element-wise from right to left (i.e., trailing dimensions first). Two dimensions are compatible when:
- They are equal, or
- One of them is 1
If the shapes are not compatible, NumPy throws a ValueError.

 Examples of Broadcasting

 Example 1: Scalar and Array


```python
import numpy as np

a = np.array([1, 2, 3])
b = 2

# b is broadcast to match the shape of a
result = a + b  # [3, 4, 5]
```

Example 2: 2D and 1D Array

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

# b is broadcast across each row of a
result = a + b
# [[11, 22, 33],
#  [14, 25, 36]]
```

 Example 3: Different Shapes


```python
a = np.array([[1], [2], [3]])  # Shape (3,1)
b = np.array([10, 20, 30])     # Shape (3,)

# b is reshaped to (1,3), then broadcast to (3,3)
result = a + b
# [[11, 21, 31],
#  [12, 22, 32],
#  [13, 23, 33]]
```

 When Broadcasting Fails


```python
a = np.array([[1, 2], [3, 4]])  # Shape (2,2)
b = np.array([1, 2, 3])         # Shape (3,)

# This will raise an error because shapes are incompatible
result = a + b  # ❌ ValueError
```

Why Is Broadcasting Useful?
- Memory-efficient: No need to replicate data manually.
- Cleaner code: Avoids loops and reshaping.
- Fast: Operations are vectorized and optimized under the hood.

4. Explain the use of the groupby() method in Pandas ?

- The groupby() method in Pandas is a powerful tool for data aggregation and analysis. It allows you to split your data into groups based on some criteria, apply a function to each group, and then combine the results.

 What Does groupby() Do?
The process follows this pattern:


```python
Split ➡ Apply ➡ Combine
```

- Split the data into groups based on column values
- Apply a function (e.g., sum, mean, count) to each group
- Combine the results into a new DataFrame or Series

Syntax



```python
df.groupby('column_name')
```

You can also group by multiple columns:

```python
df.groupby(['col1', 'col2'])
```

Example

```python
import pandas as pd

data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [50000, 55000, 60000, 62000, 70000]
}

df = pd.DataFrame(data)

# Group by Department and calculate average salary
grouped = df.groupby('Department')['Salary'].mean()

print(grouped)
```

Output:

```python
Department
HR       61000.0
IT       70000.0
Sales    52500.0
Name: Salary, dtype: float64
```

Pandas groupby() — Line-by-Line Summary
- groupby() is used to split, apply, and combine data.
- It groups data based on column values.
- You can apply functions like sum(), mean(), count(), etc.
- Syntax: df.groupby('column_name')
- You can group by multiple columns: df.groupby(['col1', 'col2'])


 Example



```python
df.groupby('Department')['Salary'].mean()
```

- Groups data by Department
- Calculates average Salary for each group

 Common Functions
- .sum() → Total per group
- .mean() → Average per group
- .count() → Number of entries per group
- .max() / .min() → Max/min per group
- .agg() → Apply multiple functions

Multiple Aggregations


```python
df.groupby('Department')['Salary'].agg(['mean', 'max', 'min'])
```

- Returns mean, max, and min salary per department

6. What are the differences between NumPy arrays and Python lists ?

- Here's a clear, line-by-line breakdown of the key differences between NumPy arrays and Python lists:

 NumPy Arrays vs Python Lists — Line-by-Line Comparison

 Data Type Consistency
- NumPy arrays: All elements must be of the same type.
- Python lists: Elements can be of mixed types (e.g., int, str, float).

 Performance
- NumPy arrays: Much faster and more memory-efficient for numerical operations.
- Python lists: Slower for large-scale computations.

 Mathematical Operations
- NumPy arrays: Support element-wise operations (e.g., array * 2).
- Python lists: You need to loop manually or use list comprehensions for math.

 Functionality
- NumPy arrays: Come with hundreds of built-in functions for stats, linear algebra, etc.
- Python lists: Basic structure with no mathematical tools.

 Shape and Dimensions
- NumPy arrays: Support multi-dimensional arrays (e.g., 2D, 3D).
- Python lists: You can nest lists, but they’re not truly multidimensional.

 Interoperability
- NumPy arrays: Easy to integrate with Pandas, SciPy, Scikit-learn, etc.
- Python lists: Less compatible with scientific libraries.

 Example Comparison


```python
import numpy as np

# NumPy array
a = np.array([1, 2, 3])
print(a * 2)  # [2, 4, 6]

# Python list
b = [1, 2, 3]
print([x * 2 for x in b])  # [2, 4, 6]
```

8.  What does the term “vectorized operation” mean in NumPy ?

- In NumPy, a vectorized operation refers to performing element-wise computations over arrays without using explicit Python loops. It’s one of the reasons NumPy is fast, efficient, and a favorite among data scientists.

 Vectorized Operation — Line-by-Line Summary
- A vectorized operation applies a function directly to entire arrays.
- Eliminates the need for manual iteration with for loops.
- Uses under-the-hood C code, making it extremely fast and memory-efficient.
- Makes code cleaner, shorter, and more readable.
- Enables element-wise arithmetic, comparison, and logical operations.

 Example: Add Arrays Without a Loop


```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

# Vectorized addition
c = a + b  # Output: [11, 22, 33]
```

No loop needed — just one line for the full array!

 Supported Operations
- Arithmetic: +, -, *, /, **
- Comparison: >, <, ==, !=
- Logical: &, |, ~

Non-Vectorized vs Vectorized


```python
# Non-vectorized (using loop)
result = []
for x, y in zip(a, b):
    result.append(x + y)

# Vectorized
result = a + b
```

- The vectorized version is 10x+ faster and more elegant.

 Benefits
- Performance boost on large datasets
- Compact and expressive syntax
- Less chance of bugs due to manual looping

 apply() vs map() — Line-by-Line Summary

 Scope of Use
- map() is used only on Series objects (i.e., one column).
- apply() can be used on both Series and DataFrames.

 Function Flexibility
- map() works best with simple functions (e.g., lambda, built-in functions, or dictionary mappings).
- apply() can handle more complex logic, especially across rows or columns.

 DataFrame Support
- map()
 does not support DataFrames directly.
- apply()
 works on entire DataFrames (can be used row-wise or column-wise).

 Axis Control
- apply() allows you to specify the axis:
- axis=0: apply function to each column
- axis=1: apply function to each row
- map() does not have axis control—it always acts element-wise on a Series.

Examples

 Using map() on a Series



```python
import pandas as pd

df = pd.DataFrame({'Score': [85, 90, 78]})

# Apply a function to a single column
df['Result'] = df['Score'].map(lambda x: 'Pass' if x >= 80 else 'Fail')
```

Using apply() on a DataFrame


```python
df['Result'] = df.apply(lambda row: 'Pass' if row['Score'] >= 80 else 'Fail', axis=1)
```

21.  Why is NumPy’s array slicing faster than Python’s list slicing ?

- NumPy’s array slicing is faster than Python’s list slicing primarily because of its under-the-hood optimizations and memory architecture designed for numerical computing. Here’s a crisp breakdown:

⚡ Why NumPy Array Slicing Is Faster — Line-by-Line Summary
- Built on C: NumPy is implemented in C, allowing low-level memory manipulation.
- Contiguous memory blocks: NumPy arrays use compact, contiguous memory, unlike Python lists which store pointers to individual objects.
- Homogeneous data types: All elements in a NumPy array have the same fixed type, enabling predictable and fast access.
- Views instead of copies: Slicing an array returns a view (i.e. a lightweight reference), not a new copy—saving memory and time.
- Vectorized backend: NumPy slicing leverages compiled routines for reading and writing blocks of data.
- Avoids overhead: Python lists require dynamic type checking and reference management for each element, adding latency.
- Optimized for math operations: NumPy arrays are aligned for CPU cache efficiency, especially when sliced in bulk operations.

 Example Comparison


```python
import numpy as np

# NumPy slicing
a = np.arange(1_000_000)
result_np = a[100:200]

# Python list slicing
b = list(range(1_000_000))
result_list = b[100:200]
```

On large datasets, NumPy will consistently outperform Python lists due to its memory-efficient architecture and compiled slicing routines.
